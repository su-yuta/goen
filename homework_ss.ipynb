## まずは実家近くの物件の一覧を取得する


```python
# 必要なライブラリをインポート
import requests # requests機能をインポートします。
from bs4 import BeautifulSoup # BeautifulSoupの機能をインポートします。
import pandas as pd
```

```python
# 対象ページの構造をまずは取得する
REQUEST_URL = "https://suumo.jp/jj/bukken/ichiran/JJ012FC001/?jj012fi20202Kbn=6&initKbn=1&displayClass=dn&disabledClass=false&ar=030&bs=020&ra=030011&rn=0395&rnTemp=0395&ekTemp=039526170&ekTjCd=&ekTjNm=&tj=0&kw=1&cn=9999999&rnek=039526170&kb=1&kt=9999999&km=1&tb=0&tt=9999999&hb=0&ht=9999999&md=3&md=4&md=5&et=9999999&et=9999999&ni=9999999&ni=9999999&fw2="
res = requests.get(REQUEST_URL) #リクエストしたデータをresに代入します。
```

```python
print(res.encoding) # res.encodingで文字エンコードを確認できます。
```

```python
soup = BeautifulSoup(res.text,"html.parser") # BeautifulSoup(解析したいデータ,解析する方法)を指定し、解析したデータをsoupに代入します。
# soup
```

## まずは物件名をとっていく


```python
soup.find_all("dd", attrs={"class":"dottable-vm"})
```

```python
# 調べてみて物件名がないものがある！！しかも、なぜか価格が２つ取得されているものもあってわけがわからん（画面上では一つしかないのに）
# 方針を変換して各物件毎のタイトルを持ってくる方針へ
soup.find_all("h2", attrs={"class":"property_unit-title"})
titles = soup.find_all("h2", attrs={"class":"property_unit-title"})
titles[0].text
```

```python
properties = soup.find_all("div", attrs={"class":"property_unit"})
property = properties[0]
property
```

```python
property_name = property.find("h2", attrs={"class":"property_unit-title"})
property_name
```

## ここまでのまとめ


```python
# ちなみにfind_all にするとResultSetになってしまうみたいで、allならTagになるそうなので、それを解消するようなcodeにしておく
# 余計な文字列を削除するようにしておく（replaceを使って）
properties = soup.find_all("div", attrs={"class":"property_unit"})
property = properties[0]
property_name = property.find("h2", attrs={"class":"property_unit-title"})
property_name = property_name.text.replace("\n","")
property_name
```

```python
# ここからは他の要素の抜き出しを行う（価格、最寄りからの距離、建物面積、間取り
# まずは値段

property_price = property.find("span", attrs={"class":"dottable-value"})
property_price = property_price.text
property_price
```

```python
# 距離、面積、間取りを確かめるための塊をまずは抽出

property_detailes = property.find_all("div", attrs={"class":"dottable-line"})
property_detailes
```

```python
# 距離

property_distance = property_detailes[-4].find_all("dd")[1].text
property_distance
```

```python
# 間取り

property_breadth = property_detailes[-3].find_all("dd")[1].text
property_breadth
```

```python
# 建物面積

property_building_area = property_detailes[-2].find("dd").text
property_building_area
```

### ここまでのまとめ


```python
properties = soup.find_all("div", attrs={"class":"property_unit"})
property = properties[0]
property_name = property.find("h2", attrs={"class":"property_unit-title"})
property_name = property_name.text.replace("\n","")

property_price = property.find("span", attrs={"class":"dottable-value"})
property_price = property_price.text

property_detailes = property.find_all("div", attrs={"class":"dottable-line"})

property_distance = property_detailes[-4].find_all("dd")[1].text
property_breadth = property_detailes[-3].find_all("dd")[1].text
property_building_area = property_detailes[-2].find("dd").text

print(property_name)
print(property_price)
print(property_distance)
print(property_breadth)
print(property_building_area)

```

```python
# 取得した情報を辞書型にまとめていく

details = {}

details["物件名"] = property_name
details["価格"] = property_price
details["最寄りからの距離"] = property_distance
details["間取り"] = property_breadth
details["建物面積"] = property_building_area

details
```
